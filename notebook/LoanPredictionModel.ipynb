```python
df = df.drop(columns=['loan_id'])
```

```python
df.head()
```

```python
df['Total_asset'] = df['bank_asset_value']+ df['residential_assets_value'] + df['luxury_assets_value'] + df['commercial_assets_value']
```

```python
df.head()
```

```python
df = df.drop(columns=['bank_asset_value','residential_assets_value','luxury_assets_value','commercial_assets_value'])
```

```python
df.columns = df.columns.str.strip()

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

print("Spaces removed from column names and string values in all rows.")
print(df.head())
```

```python
df.head()
```

```python
df['education']= df['education'].map({'Graduate':1,'Not Graduate':0})
df['self_employed']= df['self_employed'].map({'Yes':1,'No':0})
df['loan_status']= df['loan_status'].map({'Approved':1,'Rejected':0})
```

```python
df.head()
```

```python
sns.displot(df['Total_asset'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,PowerTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('loan_status', axis=1), df['loan_status'], test_size=0.2, random_state=42)
```

```python
x_train
```

```python
transformer = ColumnTransformer([
    ('scaler', StandardScaler(), [	'income_annum', 'loan_amount']),
    ('power_transformer', PowerTransformer(), ['Total_asset'])

],remainder='passthrough')
```

```python
x_train_transformed = transformer.fit_transform(x_train)
x_test_transformed = transformer.transform(x_test)
```

```python
transformed_columns = transformer.get_feature_names_out()

x_train_scaled = pd.DataFrame(x_train_transformed, columns=transformed_columns)
x_test_scaled = pd.DataFrame(x_test_transformed, columns=transformed_columns)

```

```python
np.round(x_train_scaled.describe(), 1)
```

```python
import matplotlib.pyplot as plt
```

```python
x_train_scaled.head()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['income_annum'], ax=ax1,color='red')
sns.kdeplot(x_train['loan_amount'], ax=ax1,color='green')
sns.kdeplot(x_train['Total_asset'], ax=ax1,color='blue')

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_scaled['scaler__income_annum'], ax=ax2,color='red')
sns.kdeplot(x_train_scaled['scaler__loan_amount'], ax=ax2,color='green')
sns.kdeplot(x_train_scaled['power_transformer__Total_asset'], ax=ax2,color='blue')
plt.show()
```

```python
lr = LogisticRegression(max_iter=500)
lr.fit(x_train_scaled, y_train)
```

```python
y_pred= lr.predict(x_test_scaled)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=120,
    max_depth=10,
    min_samples_split=5,
    random_state=42
)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_rf)
```

```python
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
importance = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': rf.feature_importances_
})

importance.sort_values(
    by='Importance',
    ascending=False
).head(10)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=10,
    min_samples_split=5,
    random_state=42
)
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_dt)
```
